```python
import scipy
print('correlation between errors and speeds')
# Calculate p-values of fibre speed versus rest
# I can use the paired t-test becuase each item in the dataframe has both, the fibre and the other speeds, so the order is correct, this may not be the case when comparing data across meshes
#df = df[df['meshName'] != 'DTI004']# exclude Mesh-3 i.e. DTI004
df.sort_values(by=['fileName']) # by sorting by fileName I assure that the same configurations are compared to each other since the only distinct part in the fileName defines what configuration it is
fibre_errors = np.abs(df['fibre-median-error'].values) # absolute errors for the fibre speed
degree_of_freedom = len(fibre_errors) - 1 # number of pairs - 1
print(degree_of_freedom)
conduction_speed_names = ['endo', 'sheet', 'normal']
for speed_i in range(len(conduction_speed_names)):
    speed = conduction_speed_names[speed_i]
    print(speed)
    speed_error_differences = fibre_errors - np.abs(df[speed + '-median-error'].values)
    t_score = (np.mean(speed_error_differences) - 0.)/(np.std(speed_error_differences)/len(fibre_errors)**0.5) # t = (mean_D - mu_0) / (std_D / sqrt(n)) where mu_0 is zero because we want to see if fibre error is larger
    print(scipy.stats.t.sf(abs(t_score), df=degree_of_freedom))
    print()
```

```python
import os
import re
import subprocess


def available_cpu_count():
    """ Number of available virtual or physical CPUs on this system, i.e.
    user/real as output by time(1) when called with an optimally scaling
    userspace-only program"""

    # cpuset
    # cpuset may restrict the number of *available* processors
    try:
        m = re.search(r"(?m)^Cpus_allowed:\s*(.*)$", open("/proc/self/status").read())
        if m:
            res = bin(int(m.group(1).replace(",", ""), 16)).count("1")
            if res > 0:
                return res
    except IOError:
        pass

    # Python 2.6+
    try:
        import multiprocessing

        return multiprocessing.cpu_count()
    except (ImportError, NotImplementedError):
        pass

    # https://github.com/giampaolo/psutil
    try:
        import psutil

        return psutil.cpu_count()  # psutil.NUM_CPUS on old versions
    except (ImportError, AttributeError):
        pass

    # POSIX
    try:
        res = int(os.sysconf("SC_NPROCESSORS_ONLN"))

        if res > 0:
            return res
    except (AttributeError, ValueError):
        pass

    # Windows
    try:
        res = int(os.environ["NUMBER_OF_PROCESSORS"])

        if res > 0:
            return res
    except (KeyError, ValueError):
        pass

    # jython
    try:
        from java.lang import Runtime

        runtime = Runtime.getRuntime()
        res = runtime.availableProcessors()
        if res > 0:
            return res
    except ImportError:
        pass

    # BSD
    try:
        sysctl = subprocess.Popen(["sysctl", "-n", "hw.ncpu"], stdout=subprocess.PIPE)
        scStdout = sysctl.communicate()[0]
        res = int(scStdout)

        if res > 0:
            return res
    except (OSError, ValueError):
        pass

    # Linux
    try:
        res = open("/proc/cpuinfo").read().count("processor\t:")

        if res > 0:
            return res
    except IOError:
        pass

    # Solaris
    try:
        pseudoDevices = os.listdir("/devices/pseudo/")
        res = 0
        for pd in pseudoDevices:
            if re.match(r"^cpuid@[0-9]+$", pd):
                res += 1

        if res > 0:
            return res
    except OSError:
        pass

    # Other UNIXes (heuristic)
    try:
        try:
            dmesg = open("/var/run/dmesg.boot").read()
        except IOError:
            dmesgProcess = subprocess.Popen(["dmesg"], stdout=subprocess.PIPE)
            dmesg = dmesgProcess.communicate()[0]

        res = 0
        while "\ncpu" + str(res) + ":" in dmesg:
            res += 1

        if res > 0:
            return res
    except OSError:
        pass

    raise Exception("Can not determine number of CPUs on this system")


available_cpu_count()
```
